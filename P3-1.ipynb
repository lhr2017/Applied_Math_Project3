```julia
using LinearAlgebra
using Plots
plotly()
```

```julia
function f(n,k,t)
  if k==0
    return 1/sqrt(n)
  else
    return sqrt(2/n)*cos(k*(2t+1)*pi/(2n))
  end
end
```

```julia
x=[4 3.1 2.8 2.5 2.3 2.2 2.1 2]'
n=8
C=[f(n,i,j) for i=0:(n-1), j=0:(n-1)]
y=C*x

d=0:(n-1)
t=0:.01:(n-1)
M=[f(n,i,j) for i in d, j in t]

z=M'*y
```

Part A

```julia
plot(d, x, seriestype = :scatter)
plot!(t,z)
```

Part B

```julia
m=n-2
z=M[1:m,:]'*y[1:m]
plot!(t,z)
```
